## PISA Score와 GDP 상관 관계 구하기

PISA (2006-2018, 3년 주기) 5회 모두 평가를 마친 국가들 중 상위 20개 국가와 GDP 기준 상위 50개 국가를 뽑아     
두 기준 모두 충족하는 국가의 GDP와 PISA Reading Score 의 상관 관계를 볼 예정입니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 선형 회귀,,
from sklearn import linear_model, datasets

# 한글 폰트 설정
plt.rc('font', family='NanumGothic')
%matplotlib inline
```

### 1. GDP 기준 상위 50개 국가 구하기
- `social indicator` 데이터셋 중 2011년 기준 국가별 GDP 데이터인 5a.xls를 사용합니다.      
- 불필요한 NaN 컬럼과 제목 행을 삭제 후 데이터만 컬럼을 붙여 재구성 합니다.     

```python
table5a = pd.read_excel('5a.xls')
table5a
```

### ✍ 데이터 전처리

```python
# 불필요한 컬럼 제거(값이 모두 NaN인 경우)
data = table5a

data = data.dropna(axis=1, how='all')
data
```

```python
# 실제 데이터가 있는 3행부터 추출해옴
data_without_column = data.iloc[3:]
data_without_column.info()
```

```python
# 공백 데이터를 표시하는 '…'의 경우 NaN으로 바꿔줍니다.
# 2열부터 숫자 데이터이지만 dtype이 Object이므로 float64로 바꿔줍니다.

data_without_column = data_without_column.replace({'…': np.nan})
value = data_without_column.iloc[:, 2:].astype('float64')

preprocessed = pd.concat([data_without_column.iloc[:, :2], value], axis=1)
preprocessed
```

```python
# 전처리 한 데이터에 컬럼을 붙여줍니다.

columns = ['나라', '연도', '1인당 GDP', '경제활동률-전체', '경제활동률-남성', '경제활동률-여성']
preprocessed.columns = columns
preprocessed
```

```python
preprocessed['연도'].unique()
```

```python
# 1인당 GDP를 기준으로 내림차순 정렬하여 상위 50개 데이터를 추출합니다.
all_data = preprocessed.sort_values(by='1인당 GDP', ascending=False)
top50 = all_data.head(50)
top50
```

### ✍ 결과

```python
# GDP가 높은데 경제활동률이 없는 국가의 경우 , 
# 세율이 낮아 타 국가 혹은 기업의 자금을 운용하는 산업으로 외화를 번다고 나와 있어서 일단 제외하였습니다.
# 경제활동률 데이터를 따로 쓰지 않는다면 제외하지 않고 사용할 예정입니다.

top50 = top50.dropna(axis=0)
top50
```

###  2. GDP 상위 50개국의 PISA 데이터 가져오기

- PISA의 전처리는 다른 파일에서 작성했고 해당 결과를 `pisa_gdp.csv` 파일로 가져와서 사용합니다.

```python
# 상위 50개국을 리스트로 추출해서 pisa 데이터와 비교합니다.

top50_list = top50['나라'].tolist()
top50_list
```

```python
# pisa_gdp.csv는 위 top50 리스트를 pisa에 2006년부터 2018년까지 꾸준히 참여한 국가 데이터와 비교하여 
# 공통적으로 존재하는 나라를 저장한 데이터입니다.

data = pd.read_csv('pisa_gdp.csv')
pisa_gdp = data

# pisa_gdp.columns = ['연도', '나라', 'PISA Reading Score', '순위']
pisa_gdp
pisa_gdp['Year'].unique()
```

### 3. PISA 데이터와 GDP 데이터 병합

```python
#  연도별로 분류합니다.
pisa_gdp_2006 = pisa_gdp[pisa_gdp['Year'] == 2006]
pisa_gdp_2009 = pisa_gdp[pisa_gdp['Year'] == 2009]
pisa_gdp_2012 = pisa_gdp[pisa_gdp['Year'] == 2012]
pisa_gdp_2015 = pisa_gdp[pisa_gdp['Year'] == 2015]
pisa_gdp_2018 = pisa_gdp[pisa_gdp['Year'] == 2018]

def getCountryList(df):
    return df['Country'].tolist()

pisa_gdp_2006_list = pisa_gdp_2006['Country'].tolist()
pisa_gdp_2009_list = getCountryList(pisa_gdp_2009)

print(len(pisa_gdp_2009_list))

pisa_gdp_2009
```

```python
# top50 데이터(gdp)에서 2009년 pisa 데이터에 존재하는 국가 데이터만 추출합니다.

pisa_gdp_top50 = top50[top50['나라'].isin(pisa_gdp_2009_list)]
pisa_gdp_top50.info()
```

```python
# 두 데이터를 나라 순으로 정렬하여 합칩니다. 
sorted_data = pisa_gdp_top50.sort_values(by='나라').reset_index(drop=True)
a = pisa_gdp_2009.iloc[:, 2:].reset_index(drop=True)
joined = pd.concat([sorted_data, a],  axis=1)
joined.corr()
```

```python
data_except = joined[~(joined['나라'] == 'Luxembourg')]
```

### 4. GDP와 PISA Reading Score의 관계 시각화

- Reading score가 높을수록 gdp에 어떤 영향을 주는지 보기 위해 `seaborn`의 `regplot`을 사용합니다.
- 만약 그래프가 우상향 한다면 양의 상관관계를 가진다고 할 수 있습니다. 
- `scikit-learn`의 `LinearRegression` 을 한번 사용해 보겠습니다.

```python
# 1인당 gdp와 reading Score의 관계를 확인하기 위해 Seaborn의 regplot을 그려봅니다.
# 변화를 좀더 명확하게 보기 위해 x축과 y축에 limit을 걸어줍니다.

plt.ylim(10000, 120000)
plt.xlim(450, 530)
sns.regplot(x = 'Score', y = '1인당 GDP', data = data_except)
```

```python
# sklearn 사용
# 선형 회귀 모델을 적용하기 위해 sklearn의 lineRegression을 사용해 보았습니다.
# 이 부분은 관련 지식 없이 간단하게 적용해본 test라고 보시면 됩니다.

line_fitter = linear_model.LinearRegression()
x = joined['Score']
y = joined['1인당 GDP']
line_fitter.fit(x.values.reshape(-1, 1), y)

print("기울기:", line_fitter.coef_)
plt.plot(x, line_fitter.predict(x.values.reshape(-1,1)))
plt.ylim(10000, 120000)
plt.xlim(450, 530)
sns.regplot(x = 'Score', y = '1인당 GDP', data = joined)
plt.show()
```

## 5. GDP 전체 데이터와 PISA 전체 데이터 비교
- 위 그래프는 GDP 상위 50개국 중 PISA 데이터가 존재하는 국가들을 대상으로 분석한 그래프였다면
- 아래는 해당 GDP 데이터셋과 PISA (2009) 데이터셋에 공통으로 존재하는 국가들을 추출해 비교한 그래프입니다.
- 이상치에 강한 선형 회귀 모델인 `RANSAC` 모델을 사용하였습니다.

```python
all_data = all_data.dropna(axis=0)
all_data['나라'].unique()
```

```python
all_data.loc[44, '나라'] = "Hong Kong"
all_data.loc[45, '나라'] = "Macau"
all_data.loc[133, '나라'] = "Montenegro"
all_data
```

```python
pisa_2009 = pd.read_csv("pisa_2009.csv")
pisa_2009 = pisa_2009[['Country', 'Score']]
pisa_2009.columns = ['나라', '점수']
pisa_2009.loc[28, '나라'] = "Republic of Korea"
pisa_2009.loc[43, '나라'] = "Russian Federation"
pisa_2009_list = pisa_2009['나라'].tolist()

pisa_2009
```

```python
set(pisa_2009_list) - set(all_data['나라'].tolist())
```

```python
# 데이터 병합
all_data_gdp_2009 = all_data[all_data['나라'].isin(pisa_2009_list)].sort_values(by='나라').reset_index(drop=True)
pisa_2009 = pisa_2009.sort_values(by="나라").reset_index(drop=True)
pisa_gdp_all_2009 = pd.merge(all_data_gdp_2009, pisa_2009, on="나라")
# pisa_gdp_all_2009 = pisa_gdp_all_2009.dropna(axis=0)
pisa_gdp_all_2009
```

```python
# 1인당 gdp와 reading Score의 관계를 확인하기 위해 Seaborn의 regplot을 그려봅니다.
# 변화를 좀더 명확하게 보기 위해 x축과 y축에 limit을 걸어줍니다.

# plt.ylim(10000, 120000)
# plt.xlim(450, 530)
sns.regplot(x = '점수', y = '1인당 GDP', data = pisa_gdp_all_2009)
sns.regplot(x = '점수', y = '경제활동률-전체', data = pisa_gdp_all_2009)
```

```python
pisa_gdp_all_2009.corr()
```

```python
# sklearn 사용
# RANSAC 모델을 사용해보았습니다.

X = pisa_gdp_all_2009['점수']
y = pisa_gdp_all_2009['1인당 GDP']

ransac = linear_model.RANSACRegressor()
ransac.fit(X.values.reshape(-1, 1), y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = X.values.reshape(-1, 1)
line_y_ransac = ransac.predict(line_X)

print("Ransac_coef:", ransac.estimator_.coef_)
plt.plot(
    line_X,
    line_y_ransac,
    color="red",
    label="RANSAC regressor",
)
# plt.ylim(10000, 120000)
# plt.xlim(450, 530)
sns.regplot(x = '점수', y = '1인당 GDP', data = pisa_gdp_all_2009)
plt.show()
```

## outlier 제거 후 다시 그려보기
- 기준: 
    - 1. IQR : 4분위수에서 (Q3 - Q1) 값, 이상치 범위는 (X < Q1 - 1.5 * IQR  || X > Q3 + 1.5 * IQR) 
    - 2. 3시그마 : 시그마는 표준편차이며 이상치 범위는 (X < 평균 - 3시그마 || X > 평균 + 3시그마)

```python
pisa_gdp_all_2009[['1인당 GDP', '점수']].describe()
```

### Outlier 제거 - IQR 이용

```python
fig, ax = plt.subplots(figsize=(15, 5), ncols=2)
sns.boxplot(y='1인당 GDP', data=pisa_gdp_all_2009, ax=ax[0], color='skyblue')
sns.boxplot(y='점수', data=pisa_gdp_all_2009, ax=ax[1], color='yellowgreen')
```

```python
def remove_outlier_by_IQR(df, col=[]):
    data = df
    for col_name in col:
        s = data[col_name]
        IQR = s.quantile(0.75) - s.quantile(0.25)
        data = data[s.between(s.quantile(0.25) - (1.5 * IQR), s.quantile(0.75) + (1.5 * IQR))]
    return data

data_IQR = remove_outlier_by_IQR(pisa_gdp_all_2009, ['1인당 GDP', '점수'])[['나라', '1인당 GDP', '점수']]
data_IQR.describe()
```

```python
# sklearn 사용
# RANSAC 모델을 사용해보았습니다.

X = data_IQR['점수']
y = data_IQR['1인당 GDP']

ransac = linear_model.RANSACRegressor()
ransac.fit(X.values.reshape(-1, 1), y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = X.values.reshape(-1, 1)
line_y_ransac = ransac.predict(line_X)

print("Ransac_coef:", ransac.estimator_.coef_)
plt.plot(
    line_X,
    line_y_ransac,
    color="red",
    label="RANSAC regressor",
)
# plt.ylim(10000, 120000)
# plt.xlim(450, 530)
sns.regplot(x = '점수', y = '1인당 GDP', data = data_IQR)
plt.show()
```

### Outlier 제거 - 시그마 이용
- 평균적으로 2시그마는 95%, 3시그마는 99.7% 의 데이터를 커버한다고 보면 된다.

```python
gdp_std = pisa_gdp_all_2009['1인당 GDP'].std()
score_std = pisa_gdp_all_2009['점수'].std()
```

```python
def remove_outlier_by_sigma(df, col=[], n=3):
    data = df
    for col_name in col:
        s = data[col_name]
        sigma = s.std()
        avg = s.mean()
        data = data[s.between(avg - (n * sigma), avg + (n * sigma))]
    return data

data_3sigma = remove_outlier_by_sigma(pisa_gdp_all_2009, ['1인당 GDP', '점수'], 3)[['나라', '1인당 GDP', '점수']]
data_3sigma.describe()
```

```python
data_2sigma = remove_outlier_by_sigma(pisa_gdp_all_2009, ['1인당 GDP', '점수'], 2)[['나라', '1인당 GDP', '점수']]
data_2sigma.describe()
```

```python
# sklearn 사용
# RANSAC 모델을 사용해보았습니다.

X = data_3sigma['점수']
y = data_3sigma['1인당 GDP']

ransac = linear_model.RANSACRegressor()
ransac.fit(X.values.reshape(-1, 1), y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

line_X = X.values.reshape(-1, 1)
line_y_ransac = ransac.predict(line_X)

print("Ransac_coef:", ransac.estimator_.coef_)
plt.plot(
    line_X,
    line_y_ransac,
    color="red",
    label="RANSAC regressor",
)
# plt.ylim(10000, 120000)
# plt.xlim(450, 530)
sns.regplot(x = '점수', y = '1인당 GDP', data = data_3sigma)
plt.show()
```
