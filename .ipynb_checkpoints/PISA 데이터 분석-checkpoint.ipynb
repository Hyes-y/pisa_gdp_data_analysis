# OECD 국제 학업 성취도 비교 연구(PISA) 데이터 분석

- PISA 는 OECD에서 주관하는 만 15세 학생의 읽기, 수학, 과학 소양의 성취와 추이를 국제적으로 비교하기 위한 평가입니다. 
- 평가 결과를 분석하여 우리나라 학생들의 읽기 소양(문해력)의 변화를 살펴 볼 예정입니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('pisa_2006-2018.csv')
data.info()
```

```python
data.head()
```

```python
# Subject 종류 확인
data['Subject'].unique()
```

```python
# Subject == 'Reading' 인 데이터프레임 생성
reading_scores = data[data['Subject'] == 'Reading']
reading_scores.head()
```

```python
# 결측값 확인 (해당 연도에 평가를 하지 않은 것으로 추정)
reading_scores[reading_scores['Score'].isnull()]
```

```python
reading_scores_kr = reading_scores[reading_scores['Country'] == 'South Korea']
reading_scores_kr
```

```python
# 한국의 Reading Score 차트
sns.lineplot(x="Year", y="Score", data=reading_scores_kr)
plt.show()
```

## 영역별 비교
- 'Maths', 'Science', 'Reading' 분야 별로 비교

```python
maths_scores = data[data['Subject'] == 'Maths']
maths_scores_kr = maths_scores[maths_scores['Country'] == 'South Korea']
maths_scores_kr
```

```python
science_scores = data[data['Subject'] == 'Science']
science_scores_kr = science_scores[science_scores['Country'] == 'South Korea']
science_scores_kr
```

```python
reading = reading_scores_kr[['Year', 'Score']]
reading.set_index('Year', inplace=True)

math = maths_scores_kr[['Year', 'Score']]
math.set_index('Year', inplace=True)

science = science_scores_kr[['Year', 'Score']]
science.set_index('Year', inplace=True)

sub_all = pd.concat([reading, science, math], axis=1)
sub_all.columns = ['Reading', 'Science', 'Math']
sub_all
```

```python
sns.lineplot(data=sub_all)

plt.show()
```

```python
# reading_top10 = reading_scores_notnull.sort_values(by=['Year', 'Score'], ascending=[True, False]).groupby('Year').head(10)
reading_top10 = df.sort_values(by=['Year', 'Rank'], ascending=[True, True]).groupby('Year').head(20)
reading_top10
# sns.set(rc = {'figure.figsize':(15,8)})
# sns.lineplot(x="Year", y="Score", hue="Country", data=reading_top10)
# plt.show()
```

## 2006 - 2018 PISA 평가에 꾸준히 참여한 나라들의 순위 비교

```python
dict = {}
for country in reading_top10['Country']:
    if country not in dict:
        dict[country] = 1
    else:
        dict[country] += 1
        
countries = [country for country, cnt in dict.items() if cnt == 5]
countries
```

```python
top15 = df[df['Country'].isin(countries)]
top15
```

```python
sns.set_theme(style="whitegrid")
plt.ylim(480, 560)
plt.xticks(rotation=45)
ax = sns.barplot(x="Country", y="Score", hue="Year", data=top15)
ax
```

## 각 나라의 2006년 PISA Score와 2018년 PISA Score의 증가량 비교

```python
top15_2006 = top15[top15['Year']==2006].groupby('Country')['Score'].max()
top15_2018 = top15[top15['Year']==2018].groupby('Country')['Score'].max()

top15_diff = (top15_2006 - top15_2018).reset_index(name="diff")
top15_diff
```

```python
sns.lineplot(x="Country", y="diff", data=top15_diff)
```
